## Loading libraries

```python
# Importing necessary libraries for data analysis and visualization
import pandas as pd                # Library for data manipulation and analysis
import numpy as np                 # Library for numerical operations
import seaborn as sns              # Library for statistical data visualization
import matplotlib.pyplot as plt    # Library for creating visualizations
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

### Loading the Dataset

```python
# Loading the dataset from the provided URL
url = "https://bit.ly/SupermarketsDS"
data = pd.read_csv(url)

# Displaying the first few rows of the dataset to verify successful loading
data.head()
```

```python
# Checking the datatypes of the columns in the dataset
data_types = data.dtypes

# Displaying the datatypes
print("Column Datatypes:")
print(data_types)
```

```python
# Generating a statistical summary of the dataset
summary = data.describe()

# Displaying the summary
print("Statistical Summary:")
print(summary)
```

```python
# Checking for missing values
missing_values = data.isnull().sum()
print("Missing values per column:")
print(missing_values)
```

```python
# Dropping irrelevant columns
irrelevant_columns = ['Invoice ID', 'Time']
data.drop(columns=irrelevant_columns, inplace=True)
```

## Data analysis and visualisation (Using Seaborn for visualisation)

### **Question 1**  Which product categories(Product_line) performed best.

```python
# Create a bar plot of product line vs total sales
plt.figure(figsize=(12, 6))
sns.barplot(x='Product line', y='Total', data=data)
plt.xlabel('Product Line')
plt.ylabel('Total Sales')
plt.title('Total Sales by Product Line')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Home and Lifestyle** product line stands out as the best performing category in terms of sales. It consistently achieves higher sales compared to other product categories.
Average Amount Spent: On average, customers are spending around $322.97 per purchase. This information can help guide pricing strategies and promotions.

### **Question 2:** Which branches had the highest sales and profits ?

```python
# Calculate the profit by subtracting COGS from Total
data['Profit'] = data['Total'] - data['cogs']

# Calculate the total sales and profits for each branch
branch_sales = data.groupby('Branch')['Total'].sum()
branch_profits = data.groupby('Branch')['Profit'].sum()

# Visualize the branch sales and profits using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=branch_sales.index, y=branch_sales.values, color='blue', label='Sales')
sns.barplot(x=branch_profits.index, y=branch_profits.values, color='red', label='Profits')
plt.xlabel('Branch')
plt.ylabel('Amount')
plt.title('Branch-wise Sales and Profits')
plt.legend()
plt.show()
```

Branch C shows a slightly higher level of both sales and profits compared to the other two branches, A and B. While the difference may not be substantial, it's notable that Branch C has managed to generate a slightly higher revenue margin. This could be attributed to various factors such as strategic location, targeted marketing, or specific product offerings that appeal to the customers of Branch C. However, it's important to conduct further analysis to identify the specific factors contributing to this slight difference and determine if any adjustments or strategies could be applied to the other branches to achieve similar results.

### **Question 3:** What key trends in sales in each branch are noticeable over the time.

```python
# Convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Extract the month from the 'Date' column
data['Month'] = data['Date'].dt.month

# Group the data by 'Branch' and 'Month' to calculate total sales for each month in each branch
branch_monthly_sales = data.groupby(['Branch', 'Month'])['Total'].sum().reset_index()

# Creating a line plot to visualize the trends in sales over months for each branch
plt.figure(figsize=(12, 6))
sns.lineplot(data=branch_monthly_sales, x='Month', y='Total', hue='Branch')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Key Trends in Sales Over Months for Each Branch')
plt.xticks(range(1, 5), ['Jan', 'Feb', 'Mar', 'Apr'])
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```

**February Sales Dip:**

In our analysis of the sales trends across all branches, we observed a notable decrease in sales during the month of February. This dip is consistent across all three branches and warrants further investigation to comprehend its underlying causes.

February is relatively short, and consumer behavior can vary during this month. People may have different priorities or financial goals, leading to decreased spending.

Another contributing factor could be the timing of back-to-school spending, which typically occurs in January. After the flurry of holiday shopping, many families focus on purchasing school supplies, clothing, and related items for the new academic year. This early-year spending might lead to a decrease in overall consumer spending in the subsequent month.

### **Question 4(a):** The average amount spent by customers for each branch.

```python
# Calculate the average amount spent by customers for each branch
average_spending_by_branch = data.groupby('Branch')['Total'].mean()

# Visualize the average spending by branch using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x=average_spending_by_branch.index, y=average_spending_by_branch.values, palette='Set1')
plt.xlabel('Branch')
plt.ylabel('Average Spending')
plt.title('Average Amount Spent by Customers for Each Branch')
plt.show()
```

### **Question 4(b):** The average rating for each branch.

```python
average_rating_by_branch = data.groupby('Branch')['Rating'].mean()
plt.figure(figsize=(12, 6))
sns.barplot(x=average_rating_by_branch.index, y=average_rating_by_branch.values, palette='Set1')
plt.xlabel('Branch')
plt.ylabel('Average Rating')
plt.title('Average Rating for Each Branch')
plt.ylim(5, 8) # Set y-axis limits to better visualize differences
plt.show()
```

 From the visuals, it's evident that Branch C has a slightly higher average rating compared to the other branches. This indicates that customers of Branch C have, on average, expressed a slightly higher level of satisfaction with their shopping experience compared to the other two branches. While the difference might be slight, it still suggests that Branch C is doing well in terms of providing a positive customer experience.
